```python
import pandas as pd
```

```python
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv',
                  index_col = 'Unnamed: 0')
data.head()
```

```python
data.shape
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.pairplot(data, x_vars=['TV', 'radio', 'newspaper'], y_vars= ['sales']
            ,height = 7, aspect=0.7, kind = 'reg')
```

```python
X = data[['TV','newspaper','radio']]
```

```python
y = data[['sales']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
check = lr.predict(X_test)
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test, check))
```

```python
print(metrics.mean_squared_error(y_test, check))
```

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test, check)))
```
